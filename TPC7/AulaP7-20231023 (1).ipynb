```python
#criar função 1

def criaPolinomio():
    resPol = []
    grau = int(input("Escolha o grau do polinómio: "))
    while grau >= 0:
        coe = int(input("Escolha coeficente para o grau " + str(grau) + " :"))
        if coe != 0:
            termo = (coe, grau)
            resPol.append(termo)
        grau = grau - 1

    return resPol 
```

```python
# criar função 2

def lePolinomio(fnome):
    res = []
    file = open(fnome)
    for linha in file:
        pol = []
        termos = linha.strip().split(";")
        for trm in termos:
            monomio = trm.split(",")
            if len(monomio) >= 2:
                coe = monomio[0]
                grau = monomio[1]
                pol.append((int(monomio[0]), int(monomio[1])))
        res.append(pol)
    file.close()
    print ("____________________________________")
    return res
```

```python
# criar função 3

def tabelaPolinomio(fnome):
    ordem = 1
    file = open(fnome)
    print ("-------Ordem-------||--------Polinómio -------")

    for linha in file:
        pol = []
        termos = linha.split(";") 
        for trm in termos:
            res = ""
            monomio = trm.split(",")
            if len(monomio) >= 2:
                coe = monomio[0]
                grau = monomio[1]
                if int(coe) < 0:
                    res = res + str(coe)
                else:
                    res = res + "+" + str(coe)
                if int(coe) != 0:
                    if grau != 0:
                        res = res + "x^" + str(grau) 
                pol.append(res)
        print ( "    " + str(ordem) + "     ||     " + ''.join(pol))
        ordem = ordem + 1
        print ("________________________________________")
    file.close()
```

```python
#criar função 4

def pot(a,b):
    res = 1
    while b > 0:
        res = res * a
        b = b - 1 
    return res 
    

def calcPolinomio(fnome, x, ordem):
    polinomios = lePolinomio(fnome)
    if 0 <= ordem < len(polinomios):
        polinomioselcionado = polinomios[ordem-1]
        res = 0
        for termo in polinomioselcionado:
            coe, grau = termo
            res = res + coe * pot(x, grau)
        return res
    else:
        return "Não é possível encontrar a ordem desejada no ficheiro"
```

```python
#criar função 7 

def derivarPolinomio(p):
    derivada = [] 
    for termo in p:
        if len(termo) >= 2:
            c, g = termo
            g = int(g)
            if g != 0:
                g2 = g - 1 
                derivada.append((c*g, g2))
    return derivada

def tabelaDerivadas(fnome):
    file = open(fnome)
    print(" Polinómio      ||  Derivada   ")
    for linha in file:
        print(linha)
        termos  = linha.strip().split(";")
        for t in termos:
            termo = termos[0].split(",")
            d = derivarPolinomio(termo)
        polinomio = linha.strip()
    return str(polinomio) + "      ||  " + str(d)
```

```python
#criar função 8

def somarPolinomios(fnome, ordem1, ordem2):
    polinomios = lePolinomio(fnome)
    soma = []
    ordem = 1
    if 0 <= ordem1 <= len(polinomios) and 0 <= ordem2 <= len(polinomios):
        p1 = polinomios[ordem1 - 1]
        p2 = polinomios[ordem2 - 1] 
        i = 0
        j = 0
        if len(p1) and len(p2) >= 2:

            while i < len(p1) and j < len(p2):
                c1, g1 = p1[i]
                c2, g2 = p2[j]

                if g1 == g2:
                    soma.append((c1 + c2, g1))
                    i = i + 1
                    j = j + 1
                elif g1 > g2:
                    soma.append((c1, g1))
                    i = i + 1
                else:
                    soma.append((c2, g2))
                    j = j + 1

            while i < len(p1):
                c1, g1 = p1[i]
                soma.append((c1, g1))
                i = i + 1
            
            while j < len(p2):
                c2, g2 = p2[j]
                soma.append((c2, g2))
                j = j + 1

        return "A soma dos polinómios de ordens " + str(ordem1) + " e " + str(ordem2) + " é " + str(soma)
    else:
        return "As ordens desejadas não são válidas no ficheiro"
```

```python
def lePolinomio(fnome):
    res = []
    file = open(fnome)
    for linha in file:
        pol = []
        termos = linha.strip().split(";")
        for t in termos:
            monomio = t.split(",")
            if len(monomio) >= 2:
                c = monomio[0]
                g = monomio[1]
                pol.append((int(monomio[0]), int(monomio[1])))
        res.append(pol)
    file.close()
    print ("----------------------------------------------")
    return res

lePolinomio("PolinómiosTPC.txt")
```

```python
def derivarPolinomio(p):
    derivada = [] 
    for termo in p:
        if len(termo) >= 2:
            c, g = termo
            g = int(g)
            if g != 0:
                g2 = g - 1 
                derivada.append((c*g, g2))
    return derivada

def tabelaDerivadas(fnome):
    file = open(fnome)
    print(" Polinómio      ||  Derivada   ")
    for linha in file:
        print(linha)
        termos  = linha.strip().split(";")
        print(termos)
        for t in termos:
            termo = termos[0].split(",")
            print(termo)
            d = derivarPolinomio(termo)
        polinomio = linha.strip()
    return str(polinomio) + "      ||  " + str(d)


tabelaDerivadas("PolinómiosTPC.txt")
```

```python
def pot(a,b):
    res = 1
    while b > 0:
        res = res * a
        b = b - 1 
    return res     

def calcPolinomio(fnome, x, ordem):
    polinomios = lePolinomio(fnome)
    if 0 <= ordem < len(polinomios):
        polinomioselcionado = polinomios[ordem-1]
        res = 0
        for termo in polinomioselcionado:
            coe, grau = termo
            res = res + coe * pot(x, grau)
        return res
    else:
        return "A ordem desejada não é válida no ficheiro"
```
